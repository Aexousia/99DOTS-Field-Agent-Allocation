```python
import numpy as np
import queue
import copy
import matplotlib.pyplot as plt
import pylab as plt

final1=[]
final2=[]

for pc in range(100,2000):
    task1=[]
    task2=[]
    
    for pcm in range(5):
        list1=[]
        value=[]
        value1=[]
        sum3=0
        sum4=0
        
        total_time = 5000        
        ST_rate = 1/4 
        num_total = pc
        m= pc/100       #this might lead to a trouble wrt costant paths
        IAT_rate=1/m

        thresh = 3
        addh = 0.55
        P_miss_Add = 0.78
        P_miss_noAdd = 0.93
        p_th=0.847
        abc1=0

        qu = queue.Queue()
        curr= None

        wait_time = []
        server_busy = False
        d=0

        P_miss = []
        P_add = []
        miss_days_1=[]
        miss_days = []
        Simul_me= []
        len1=[]
    
        for k in range(400000):
            miss_days_1.append(0)
            miss_days.append(0)
            wait_time.append(0)
            
        for k in range(total_time):
            Simul_me=[]
            a1=np.random.poisson(1/IAT_rate)
            a2=np.random.poisson(1/ST_rate)
            for i in range(a1):
                list1.append(0)
                
            for i in range(a1):
                temp = np.random.binomial(a1,addh)/a1
                P_add.append(temp)
                P_miss.append(P_miss_Add*P_add[i]+ P_miss_noAdd*(1-P_add[i]))
                
            l=0
            for i in range(len(list1)):
                if(list1[i]<100):
                    l=l+1
                    
            for i in range(len(list1)):
                if(list1[i]<100 and l>0):
                    temp = np.random.binomial(l,addh)/l  
                    P_add[i]= temp
                    P_miss[i]=P_miss_Add*P_add[i]+ P_miss_noAdd*(1-P_add[i])
                    
            for i in range(len(list1)):
                list1[i]=list1[i]+1
            for i in range(len(list1)):
                if(P_miss[i]> p_th):
                    if(list1[i]<100):
                        if(miss_days_1[i]>=1):
                            miss_days[i]= miss_days_1[i]+1
                        if(miss_days_1[i]==0):
                            miss_days[i]=1
                else:
                    if(list1[i]<100):
                        miss_days[i]=0
                        
            for i in range(len(list1)):
                if(miss_days[i]>thresh and list1[i]<100):
                    Simul_me.append(i)
            a= len(Simul_me)
            for j in range(a):
                if Simul_me[j] not in qu.queue:
                    qu.put(Simul_me[j])
                    
            b3=0
            b=qu.qsize()
            cb=b
            b2=a2
            
            for j in range(a2):
                if(b>0 and b2>0):
                    curr= qu.get()
                    miss_days[curr]=0
                    P_add[curr]=1.5*P_add[curr]
                    b=b-1
                    b2=b2-1
                    d=d+1
                    
            for item in list(qu.queue):
                if(list1[item]<100):
                    wait_time[item]=wait_time[item]+1
                    
            for i in range(len(list1)):
                miss_days_1[i]=miss_days[i]
                
            for i in range(len(list1)):
                miss_days[i]=0
                
            abc1=abc1+b3
            sum1=0
            
            for i in range(len(list1)):
                if(list1[i]<100):
                    sum1=sum1+wait_time[i]
            if(abc1!=0):
                value.append(sum1/abc1)
            if(abc1==0):
                value.append(sum1)
            value1.append(cb)
        
        for kl in range(len(value)-1000,len(value)-1):
            sum3=sum3+value[kl]
        sum3=sum3/1000
        task1.append(sum3)    
         
        for kl in range(len(value1)-1000,len(value1)-1):
            sum4=sum4+value1[kl]
        sum4=sum4/1000
        task2.append(sum4)
        
    sum5=0
    sum6=0
    for i in range(len(task1)):
        sum5=sum5+task1[i]
    final1.append(sum5/5)
    
    for i in range(len(task2)):
        sum6=sum6+task2[i]
    final2.append(sum6/5)
    print(pc)
        
plt.plot([i+1 for i in range(0,len(final1))], final1)   #value1 is queue length and value is avg wating time
plt.ylabel("Avg waiting time")
plt.xlabel("No. of days")
plt.show() 


plt.plot([i+1 for i in range(0,len(final2))], final2)   #value1 is queue length and value is avg wating time
plt.ylabel("Avg length of the queue")
plt.xlabel("No. of days")
plt.show()                    
                    
                
```
